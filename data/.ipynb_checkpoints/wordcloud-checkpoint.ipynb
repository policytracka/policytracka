```python
from wordcloud import WordCloud # ใช้ทำ Word Cloud
import pandas as pd # ใช้โหลดข้อมูลและสร้างข้อมูล
import matplotlib.pyplot as plt # ใช้ Visualize Word Cloud
from pythainlp.tokenize import word_tokenize # เป็นตัวตัดคำของภาษาไทย
from pythainlp.corpus import thai_stopwords # เป็นคลัง Stop Words ของภาษาไทย 
import nltk
```

```python
df = pd.read_csv('./datasets/Open source [WeVis-Promise Tracker] Data - promise.csv') # โหลดข้อมูลไฟล์ Excel ของเรา
```

```python
df.head()
```

```python
text = [x.replace('\n', '') for x in df.promiseTitle]
text = ''.join(text)
```

```python
wt = word_tokenize(text, engine='newmm')
```

```python
# exclusion
exclusion = [' ', '  ', '(', ')', 'ปี', 'บาท']
```

```python
wt = [x for x in wt if x not in exclusion]
```

```python
# wts = [f'{a}<>{b}' for a, b in list(nltk.bigrams(wt))] # this can be use for bigram word cloud but it won't be used here
```

```python
# !wget "https://thaifonts.net/downloadfiles/zippedfontstyles/p/r/prompt-medium.zip?st=NHEQXo4TRt3y9m7pYt3JtQ&e=1678385497"
```

```python
path = 'Prompt Medium.ttf' # ตั้ง path ไปหา font ที่เราต้องการใช้แสดงผล
wordcloud = WordCloud(
                      font_path=path, # font ที่เราต้องการใช้ในการแสดงผล เราเลือกใช้ THSarabunNew 
                      stopwords=thai_stopwords(), # stop words ที่ใช้ซึ่งจะโดนตัดออกและไม่แสดงบน words cloud 
                      relative_scaling=0.3,
                      min_font_size=1,
                      background_color = "white",
                      width=1024,
                      height=768,
                      max_font_size=140,
                      max_words=400, # จำนวนคำที่เราต้องการจะแสดงใน Word Cloud
                      colormap='plasma', 
                      scale=3,
                      font_step=4,
                      collocations=False,
                      regexp=r"[ก-๙a-zA-Z']+", # Regular expression to split the input text into token
                      margin=2
                      ).generate(' '.join(wt)) # input คำที่เราตัดเข้าไปจากตัวแปร wt ในรูปแบบ string
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 12))
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")
fig.savefig('./savefile.png')
fig.show()
```
